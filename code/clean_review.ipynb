```python
# Import the necessary libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pandas as pd
import gensim
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('vader_lexicon')
```

```python
df_closed = pd.read_csv("data_closed2.csv")
df_closed
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
for idx in df_closed.index:
    review = df_closed.at[idx,"text_clean"]
    test_subset = review.split(" ")
    sid = SentimentIntensityAnalyzer()
    pos_word_list=[]
    neg_word_list=[]

    for word in test_subset:
        if (sid.polarity_scores(word)['compound']) > 0.5:
            pos_word_list.append(word)
        elif (sid.polarity_scores(word)['compound']) <-0.5:
            neg_word_list.append(word)
  
    df_closed.at[idx,"positive"] = " ".join(pos_word_list)
    df_closed.at[idx,"negative"] = " ".join(neg_word_list)
```

```python
df_closed
```

```python
aggregation_functions = {"name":"first", 'stars_y': 'mean', "text_clean":' '.join ,'positive': ' '.join, 'negative':  ' '.join}
df_closed_new = df_closed.groupby(df_closed['business_id']).aggregate(aggregation_functions)
```

```python
try:
    df_closed_new["positive"] = df_closed_new["positive"].apply(lambda x:x.split())
    df_closed_new["negative"] = df_closed_new["negative"].apply(lambda x:x.split())
except:
    pass
```

```python
df_closed_new
```

```python
df_closed_new["positive_count"] = df_closed_new["positive"].apply(lambda x:len(x))
df_closed_new["negative_count"] = df_closed_new["negative"].apply(lambda x:len(x))
df_closed_new["idx"] = range(0,len(df_closed_new))
df_closed_new.set_index("idx",inplace=True)
df_closed_new
```
