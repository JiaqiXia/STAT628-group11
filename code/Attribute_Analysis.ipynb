# Attribute Analysis

```python
import pandas as pd
import ast
import json
import seaborn as sns
```

```python
data = pd.read_csv("data_cleaned.csv", index_col = 0)
data
```

### ast.literal_eval有问题跑不出来

```python
data = data.loc[~(pd.isna(data["attributes"])), :]
try:
    data["attributes"] = data["attributes"].apply(ast.literal_eval)
except:
    pass
data[["attributes"]].iloc[0,0]
```

```python
attributes_list = []
for attribute in data["attributes"]:
    attributes_list.extend(attribute.keys())
attr_dict = pd.Series(attributes_list).value_counts().to_dict()
attr_dict
```

### We choose attributes that are provided by at least 90% of the businesses

```python
attr_list = [k for (k,v) in attr_dict.items() if v >=len(data)*0.9]
attr_list
```

```python
df_attributes = pd.json_normalize(data['attributes'])
```

### 最开始合并business和review的时候两个数据里都有star这个column，但是不一样。这里用了business的star

```python
df_attributes = df_attributes[attr_list]
df_attributes["business_id"] = data["business_id"]
df_attributes["stars_x"] = data["stars_x"]
```
